```python
import pandas as pd
import numpy as np
import requests
import urllib3

from bs4 import BeautifulSoup
```

```python
data = pd.read_excel('remaining.xlsx')
```

```python
package_names = pd.unique(data.Name).tolist()
pypi_url_names = data.PYPI_URL
```

```python
http = urllib3.PoolManager()
```

```python
urllib3.disable_warnings()
successful_files = 0
unsuccessful_files = 0
log_file = pd.DataFrame({'Name':[], 'PYPI_URL':[], 'Maintainer':[], 'Poss_Github_Page':[]})
for packs in range(len(pypi_url_names)):
    filename = "from_PYPI/{}_PYPI.txt".format(package_names[packs])
    url = pypi_url_names[packs]
    r = http.request('GET',url)
    if r.status == 200:
        html_content = requests.get(url).text
        split_maintainer = html_content.split('<a href="/user/')
        get_information = split_maintainer[1].split('/')
        content = get_information[0]
        
        github_page = 'https://github.com/' + content + '/' + package_names[packs]

        new_row = {'Name':package_names[packs], 'PYPI_URL':url, 'Maintainer':content, 'Poss_Github_Page':github_page}
        log_file = log_file.append(new_row, ignore_index=True)
        successful_files = successful_files + 1
        
    else:
        new_row = {'Name':package_names[packs], 'PYPI_URL':url, 'Maintainer':'Unsuccessful', 'Poss_Github_Page':'Unsuccessful'}
        log_file = log_file.append(new_row, ignore_index=True)
        unsuccessful_files = unsuccessful_files + 1
```
